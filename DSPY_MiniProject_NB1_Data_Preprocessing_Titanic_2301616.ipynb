## Handling Missing Values

Different strategies are applied:

- **Age**: Filled with mean value (since it is numerical)
- **Embarked**: Filled with mode (most frequent category)
- **Cabin**: Dropped due to excessive missing values

This ensures dataset completeness.

```python
# Fill Age with mean
df['Age'] = df['Age'].fillna(df['Age'].mean())

# Fill Embarked with mode
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Drop Cabin safely
df.drop(columns=['Cabin'], inplace=True, errors='ignore')
```

## Data Cleaning

Data cleaning involves:

- Removing duplicate records to avoid bias
- Converting categorical data into numerical format

Here:
- 'Sex' column is converted:
  - male → 0
  - female → 1

This makes the data suitable for machine learning models.

```python
# Remove duplicates
df.drop_duplicates(inplace=True)

# Convert categorical to numeric
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```

## Data Filtering and Feature Selection

Only relevant columns are selected for analysis:

- Survived
- Pclass
- Sex
- Age
- Fare
- Embarked

Irrelevant columns like Name, Ticket, and PassengerId are removed.

This reduces noise and improves efficiency.

```python
# Keep only relevant columns
df = df[['Survived', 'Pclass', 'Sex', 'Age', 'Fare', 'Embarked']]
df.head()
```
