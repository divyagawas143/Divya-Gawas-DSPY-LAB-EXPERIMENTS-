# Titanic Dataset - Exploratory Data Analysis (EDA)

## Objective
The objective of this notebook is to:
- Perform exploratory data analysis on the cleaned dataset
- Use graphical and non-graphical techniques
- Identify patterns, relationships, and trends
- Draw meaningful conclusions from the data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/titanic_cleaned.csv")
df.head()
```

## NON-GRAPHICAL EDA
### Analysis 1: Age Distribution (Non-Graphical)

- **Type**: Univariate Analysis  
- **Objective**: To understand the central tendency and spread of Age  

### Observation:
- Mean age ≈ (check output)
- Median age ≈ (check output)
- Presence of variation in age values

### Conclusion:
Most passengers fall within a certain age range, indicating a diverse age distribution.

```python
df['Age'].mean(), df['Age'].median(), df['Age'].std()
```

### Analysis 2: Survival vs Gender (Non-Graphical)

- **Type**: Multivariate Analysis  
- **Objective**: To compare survival rate across genders  

### Observation:
- Female survival rate is higher than male
- Males have lower survival proportion

### Conclusion:
Gender plays a significant role in survival; females had a higher chance of survival.

```python
df.groupby('Sex')['Survived'].mean()
```

## GRAPHICAL EDA
### Analysis 3: Age Distribution (Graphical)

- **Type**: Univariate Graphical Analysis  
- **Objective**: To visualize distribution of Age  

### Observation:
- Distribution is slightly right-skewed
- Majority of passengers are between 20–40 years

### Conclusion:
The dataset contains a higher proportion of young adults.

```python
plt.hist(df['Age'], bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

### Analysis 4: Survival vs Gender (Graphical)

- **Type**: Multivariate Graphical Analysis  
- **Objective**: To visually compare survival distribution by gender  

### Observation:
- Females show higher survival count
- Males show higher non-survival count

### Conclusion:
Visualization confirms that females had better survival chances compared to males.

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.show()
```

### Analysis 5: Passenger Class vs Survival

- **Type**: Multivariate Graphical Analysis  
- **Objective**: To analyze how passenger class affects survival  

### Observation:
- First class passengers have higher survival rate
- Third class passengers have lower survival rate

### Conclusion:
Socio-economic status (Pclass) significantly influenced survival chances.

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

### Analysis 6: Fare vs Survival

- **Type**: Multivariate Graphical Analysis  
- **Objective**: To analyze relationship between Fare and Survival  

### Observation:
- Higher fare passengers tend to survive more
- Lower fare passengers show higher death rate

### Conclusion:
Passengers who paid higher fares (likely higher class) had better survival chances.

```python
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title("Fare vs Survival")
plt.show()
```

## Final Conclusion

- Gender is a strong factor influencing survival
- Passenger class plays a major role in survival probability
- Age distribution shows majority are young adults
- Higher fare correlates with higher survival

Overall, socio-economic and demographic factors significantly influenced survival outcomes in the Titanic dataset.
